```python
import pandas as pd
from modeling_gpt import GPTInference
```

```python
path = '/home/ramprasad.sa/qf_sd_summarization/datasets/postprocessed/question_article_pair.csv'
df = pd.read_csv(path)
```

```python
df
```

```python
gpt_inference = GPTInference()
```

```python
import time
import tiktoken
tokenizer = tiktoken.encoding_for_model("gpt-3.5-turbo-16k-0613")
instruction = "Please create a summary for the following question only using information in the article given above"
summaries = []
for idx, row in df.iterrows():
    query = row['Query']
    article = row['Article']
    article = tokenizer.decode(tokenizer.encode(article)[:15000])
    prompt = f'Article: {article}\n{instruction}\n{query}\nSummary:'
    summary = gpt_inference.get_chatgpt_response(prompt)
    summaries += [summary]
    time.sleep(4)
```
