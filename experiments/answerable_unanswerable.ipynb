```python
from datasets import load_dataset
import pandas as pd
oasum_dataset = load_dataset("kqsong/OASum", split = "test", cache_dir='/scratch/ramprasad.sa/huggingface_datasets')
natural_questions =load_dataset("natural_questions", split = "validation", cache_dir='/scratch/ramprasad.sa/huggingface_datasets')
```

```python
oasum_titles = ['Best Buddies International', 'James T. Brady', 'Israeli Jews']
```

```python

sample_dataset = {
    
}

for datum in oasum_dataset:
    title = datum['title']
    doc = datum['document']
    aspect = datum['aspect']
    aspect_sents = datum['aspect_sents']
    summary = datum['summary']
    if title in oasum_titles:
        if title not in sample_dataset:
            sample_dataset[title] = {'original_doc': doc, 'excluded_docs': [], 'aspects': [], 'aspect_sents': []}
        sample_dataset[title]['aspects'].append(aspect)
        aspect_sents = [doc[each] for each in aspect_sents]
        sample_dataset[title]['aspect_sents'].append(aspect_sents)
        excluded_doc = [each for each in doc if each not in aspect_sents]
        sample_dataset[title]['excluded_docs'].append(excluded_doc)
        
```

```python
custom_questions = {'Israeli Jews': {
                                'History':    [['Generate a summary of the population of Israeli Jews as founded by Israel\'s Central Bureau of Statistics',
                                             ], 
                                             
                                             ['Generate a summary of the origins of Israeli Jews according to the Hebrew Bible',
                                             'Generate a summary about the 1948 Arab–Israeli War']],
    
                                
            
                                },
            'James T. Brady': {
                              'Early life' : [['Summarize James T. Brady\'s prominent cases and achievements in criminal law',
                               ],
                              ['Summarize James T. Brady\'s childhood']],
             
                              'Personal life and death': [['Summarize James T. Brady\'s political affiliations, social activities, and personal choices'
                              ],
                              ['Summarize how James T. Brady\'s died']]
                            },
             
             'Best Buddies International': {
                                            'Fundraising and events ; Best Buddies Challenge': [['What are the programs Best Buddies offers? Describe them',
                                             'Give me details about the Best Buddies x Rosetta Bakery partnership',
                                            ], 
                                             
                                             ['What is Best Buddies Challenge?']],
    
                                             'Fundraising and events ; Best Buddies Ventures' : [['What are the formal programs Best Buddies offers? Describe them',
                                              'What is Best Buddies Challenge?',],
                                             ['Summarize Best Buddies x Rosetta Bakery partnership']],
                                            
            
                                        }
            }
```

```python
articles = []
held_out_article = []
questions = []
faithful_query = []

for title, aspect_info in custom_questions.items():
    aspect_sents = sample_dataset[title]['aspect_sents']
    aspect_sents = [' '.join(each) for each in aspect_sents]
    excluded_docs = sample_dataset[title]['excluded_docs']
    excluded_docs = [' '.join(each) for each in excluded_docs]
    print(aspect_info)
    for aspect_name, aspect_questions in aspect_info.items():
        print(aspect_name)
        aspect_idx = sample_dataset[title]['aspects'].index(aspect_name)
        faithful_doc_questions = aspect_questions[0]
        faithful_aspect_questions = aspect_questions[1]
        
        articles += [aspect_sents[aspect_idx]] * len(faithful_doc_questions)
        questions += faithful_doc_questions
        held_out_article += [excluded_docs[aspect_idx]] * len(faithful_doc_questions)
        faithful_query += ['no'] * len(faithful_doc_questions)
        
        articles += [excluded_docs[aspect_idx]] * len(faithful_aspect_questions)
        questions += faithful_aspect_questions
        held_out_article += [aspect_sents[aspect_idx]] * len(faithful_aspect_questions)
        faithful_query += ['no'] * len(faithful_aspect_questions)
    if len(articles) > 500:
        break
```

```python
from collections import Counter
articles = []
held_out_article = []
questions = []
faithful_query = []

for natural_question_datum in natural_questions:
    question = natural_question_datum['question']['text']
    document = natural_question_datum['document']
    document = ' '.join([each for each in document['tokens']['token'] if '<' not in each])
    annotations = natural_question_datum['annotations']
    long_answer_candidates = [each['candidate_index'] for each in annotations['long_answer']]
    
    long_answer_index = Counter(long_answer_candidates).most_common()[0][0]
    if long_answer_index == -1:
        articles.append(document)
        questions.append(question)
        held_out_article.append(None)
        faithful_query.append('no')
    if len(articles) > 500:
        break
```

```python
import random 
id = [random.getrandbits(128) for i in range(0, len(articles))]
```

```python
df_dict = {'Article': articles,
          'Held_Out_Article':held_out_article,
          'Query': questions,
          'Faithful_Query': faithful_query,
          'id': id}
df = pd.DataFrame(df_dict)
```

```python
from modeling_gpt import GPTInference
gpt_inference = GPTInference()
import time
import tiktoken
tokenizer = tiktoken.encoding_for_model("gpt-3.5-turbo-16k-0613")
instruction = "Please create a summary for the following question only using information in the article given above"
summaries = []
for idx, row in df[:100].iterrows():
    query = row['Query']
    article = row['Article']
    article = tokenizer.decode(tokenizer.encode(article)[:15000])
    prompt = f'Article: {article}\n{instruction}\n{query}\nSummary:'
    summary = gpt_inference.get_chatgpt_response(prompt)
    summaries += [summary]
    time.sleep(4)
```

```python
df = df[:100]
df['Summary'] = summaries
df['system_id'] = ['gpt3.5'] * len(summaries)
```
