```python
from datasets import load_dataset
# from datasets import load_dataset
import random 
import pandas as pd
dataset = load_dataset("drop", split = "validation", cache_dir='/scratch/ramprasad.sa/huggingface_datasets')
```

```python
dataset[0]
```

```python
import random 

df_dict = {'id': [],
          'Article': [],
           'Query': [],
           'Faithful_Query': [],
           'Complex_Query': [],
           
          }
for datum in dataset:
    id = random.getrandbits(128)
    passage = datum['passage']
    question = datum['question']
#     print(datum)
    answer = ', '.join(datum['answers_spans']['spans'])
    df_dict['id'] += [id]
    df_dict['Article'] += [passage]
    df_dict['Query'] += [question]
    df_dict['Faithful_Query'] += ['yes']
    df_dict['Complex_Query'] += ['yes']
    
    
```

```python
df = pd.DataFrame(df_dict)
```
